```python
%%sql 
SELECT Employees.FirstName, SUM(Products.Price) 
FROM Employees  
LEFT JOIN Orders on Employees.EmployeeID = Orders.EmployeeID 
LEFT JOIN OrderDetails on Orders.OrderID = OrderDetails.OrderID 
LEFT JOIN Products on OrderDetails.ProductID = Products.ProductID GROUP BY FirstName;
```

```python
%%sql
SELECT em.FirstName, SUM(p.Price) toplam_satis 
FROM Orders o 
LEFT JOIN OrderDetails od on o.OrderID = od.OrderID 
LEFT JOIN Products p on od.ProductID = p.ProductID 
LEFT JOIN Employees em on o.EmployeeID = em.EmployeeID 
GROUP BY o.EmployeeID 
ORDER BY toplam_satis DESC;
```

```python
#ALMANYADAN VERİLEN SİPARİŞLERİN KATEGORİLERİNE GÖRE ORTALAMA FİYATLARI
```

```python
%%sql
SELECT ca.CategoryName, AVG(p.Price) 
FROM Orders o 
LEFT JOIN OrderDetails od on o.OrderID = od.OrderID 
LEFT JOIN Products p on od.ProductID = p.ProductID 
LEFT JOIN Categories ca on p.CategoryID = ca.CategoryID 
LEFT JOIN Customers cu on o.CustomerID = cu.CustomerID 
WHERE cu.Country = 'Germany' GROUP BY CategoryName;
```

```python
#ALMANYADAN YA DA USA DAN VERİLEN SİPARİŞLERİN KATEGORİLERİNE GÖRE ORTALAMA FİYATLARI
```

```python
%%sql
SELECT ca.CategoryName, AVG(p.Price) 
FROM Orders o 
LEFT JOIN OrderDetails od on o.OrderID = od.OrderID 
LEFT JOIN Products p on od.ProductID = p.ProductID 
LEFT JOIN Categories ca on p.CategoryID = ca.CategoryID 
LEFT JOIN Customers cu on o.CustomerID = cu.CustomerID 
WHERE cu.Country = 'Germany' OR cu.Country = 'USA' GROUP BY CategoryName;
```

```python
# Müşterilerin 1997 yılına ait siparişlerinin maksimum miktarlarını bulup müşterilerin isimleri ile birlikte büyükten küçüğe sıralayınız.
```

```python
%%sql
SELECT Customers.CustomerName, MAX(Products.Price)
FROM Orders 
LEFT JOIN Customers on Orders.CustomerID = Customers.CustomerID 
LEFT JOIN OrderDetails on Orders.OrderID = OrderDetails.OrderID
LEFT JOIN Products on OrderDetails.ProductID = Products.ProductID
WHERE OrderDate like '%1997%' GROUP BY CustomerName ORDER BY Products.Price DESC;
```

```python
%%sql
SELECT CustomerName, MAX(Price) maks 
FROM Products 
LEFT JOIN OrderDetails on Products.ProductID = OrderDetails.ProductID 
LEFT JOIN Orders on OrderDetails.OrderID = Orders.OrderID 
LEFT JOIN Customers on Customers.CustomerID = Orders.CustomerID 
WHERE OrderDate like '%1997%' 
GROUP BY CustomerName ORDER BY maks DESC;
```

```python
#Sipariş yılı 1997 olan siparişleri alan çalışanları sipariş alma sayılarına göre sıralayınız.
```

```python
%%sql
SELECT COUNT(*) n, Employees.FirstName 
FROM Orders 
LEFT JOIN Employees on Orders.EmployeeID = Employees.EmployeeID 
WHERE OrderDate like '%1997%' 
GROUP BY Employees.FirstName 
ORDER by n DESC;
```
